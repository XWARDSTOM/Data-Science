```python
# first and 5th visit of every day every member
df.sort_values(by='DateTime',ascending=True,inplace=True)
df['visit_num'] = df.groupby(['Member Code']).cumcount()+1
first_and_fifth_visit = df[(df['visit_num']==1) | (df['visit_num']==5)]
first_and_fifth_visit.sort_values('Member Code',ascending=True).groupby('Member Code').head()
```

```python
 #avg number of page visits daily for member
avg_visits_df = df.groupby(['Member Code','DateTime'])['visit_num'].mean()
avg_visits_df
```

```python
#stat difference between duration of visits on deposits and withdrawal page
#Null Hypothesis - 2 independent samples(duration of visits on deposits and withdrawal page) are identical by default 
#ie: there is no statistical difference between duration on deposits and withdrawal page

deposits_duration = df[df['Page']=='Deposit']['Duration'].values
withdrawal_duration = df[df['Page']=='Withdraw']['Duration'].values
```

```python
# If we observe a large p-value, for example larger than 0.05 then we cannot reject the null hypothesis of identical average scores
N = len(df)
a = deposits_duration
b = withdrawal_duration
t, p = stats.ttest_ind(a,b)
print("t = " + str(t))
print("p = " + str(p))
# we can reject Null Hypothesis since p=0.60 means, there is significant statistical difference duration on deposits and withdrawal page
```

```python
# probability of deposit>>>withdraw and Home>>>deposit on the same day for a user
```

```python
kk = pd.DataFrame(df.groupby(['Member Code','DateTime'],as_index=False)['visit_num'].sum()) 
pp = pd.merge(kk.drop(columns=['visit_num']),df[['Member Code','DateTime','Page']],on=['Member Code','DateTime'],how='inner').set_index(['Member Code','DateTime'])
pp.head() 
```

```python
def calcuate_probability(newdf):
    visit_order = []
    home_deposit = []

    for j,i in enumerate(new_df.iterrows()):
        count = 0
        page = i[1]['Page']
        
        visit_order.append(page)
        
        if len(visit_order)>2:

            if page == 'Deposit'and visit_order[-2]=='Home':  
                
                home_deposit.append(page)
                count += 1


        prob = len(home_deposit)/(j+1)

    print('Probability of Home>>>deposit ',round(prob*100,2))


```

```python
def calcuate_probab(newdf):
    visit_order = []
    home_deposit = []

    for j,i in enumerate(new_df.iterrows()):
        count = 0
        page = i[1]['Page']
        
        visit_order.append(page)
        
        if len(visit_order)>2:

            if page == 'Deposit'and visit_order[-2]=='Withdraw':  
                
                home_deposit.append(page)
                count += 1


        prob = len(home_deposit)/(j+1)

    print('Probability of Deposit>>>Withdraw ',round(prob*100,2))
```

```python
for member_code, new_df in pp.groupby(level=0):
    print('Member Code : ',member_code)
    calcuate_probability(new_df)
    calcuate_probab(new_df)
```
